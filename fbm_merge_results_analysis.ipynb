```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rr = pd.read_csv('./results/merge_experiment_fbm_results.csv')
rr
```

```python
rr['hurst_exponent'] = round(rr['hurst_exponent'], 2)
```

```python
groups = rr.groupby(by=['hurst_exponent', 'algorithm', 'chunk_size'])
```

```python
means = groups.agg('mean')['merge_mean'].unstack('chunk_size')
means
```

```python
hursts = [0.25, 0.35, 0.45, 0.55, 0.65, 0.75]
fig = plt.figure(figsize=(8,8))
rows, cols = 3, 2
ax = None
for x in range(rows * cols):
    ax = plt.subplot(3,2,x+1,sharex=ax)
    if x%2:
        ax.yaxis.tick_right()
        ax.yaxis.set_label_position('right')
    hurst = hursts[x]
    df_to_plot = means.groupby(level=0).get_group(hurst).transpose()
    plt.plot(df_to_plot);
    plt.yscale('log')
    plt.text(.1, .9, f'h={hurst}', horizontalalignment='center', transform=ax.transAxes)
    plt.tight_layout()
    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0), useMathText=True)
    plt.xlabel("Subgraph size (vertices)")
    plt.ylabel('time (s)')
handles = ax.get_lines()
labels = ['Binary Search Tree HVG', 'Dual Tree HVG']
lgd = fig.legend(handles, labels, loc='lower center', bbox_to_anchor=(0.5,-0.075))
# txt = plt.suptitle('Mean merge times, for subgraphs of HVGs containing 1M vertices\ngenerated by fractional Brownian motions with varying Hurst index $h$', fontsize=14, x=0.5, y=1.08);
plt.savefig('fbm_merge_times.pdf', bbox_extra_artists=(lgd,ax), bbox_inches='tight');
plt.show();
```
